```python
import torch
import torch.nn as nn
import torchvision.transforms as transforms
from torchview import draw_graph

# Define the SRCNN model
class SRCNN(nn.Module):
    def __init__(self):
        super(SRCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 64, kernel_size=9, padding=4)  # Input: 3 channels (RGB), Output: 64 channels
        self.conv2 = nn.Conv2d(64, 32, kernel_size=5, padding=2) # Input: 64 channels, Output: 32 channels
        self.conv3 = nn.Conv2d(32, 3, kernel_size=5, padding=2)  # Input: 32 channels, Output: 3 channels (RGB)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = self.relu(self.conv2(x))
        x = self.conv3(x)
        return x

# Initialize the SRCNN model
model = SRCNN()

# Draw the model graph with specified input size (for example: batch_size=1, channels=3, height=33, width=33)
model_graph = draw_graph(model, input_size=(1, 3, 33, 33), expand_nested=True)

# Visualize the graph
model_graph.visual_graph
```
